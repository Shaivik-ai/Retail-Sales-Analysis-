```python
import pandas as pd

df = pd.read_csv("supermarket_sales.csv")
df.columns = df.columns.str.replace(" ", "_").str.upper()
df.dropna(subset=["PRODUCT_LINE", "TOTAL", "COGS", "GROSS_INCOME"], inplace=True)
```

```python
import numpy as np
df["INVENTORY_DAYS"] = np.random.randint(10, 180, size=len(df))
df["PROFIT_MARGIN"] = df["GROSS_INCOME"] / df["COGS"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x="INVENTORY_DAYS", y="PROFIT_MARGIN", data=df)
plt.title("Inventory Days vs Profit Margin")
plt.show()
```
